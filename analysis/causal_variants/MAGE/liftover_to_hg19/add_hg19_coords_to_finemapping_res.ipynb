```python
HG19_VCF_PATH = "/data/yosef3/scratch/ruchir/data/MAGE/MAGE.v1.0.data/QTL_results/eQTL_results/eQTL_finemapping_results/hg19/significantVariants.hg19.vcf"
```

```python
def parse_info_field(info_str: str) -> dict:
    info_dict = {}
    for attr in info_str.split(";"):
        if "=" not in attr:
            continue
        k, v = attr.split("=")
        info_dict[k] = v
    return info_dict
```

```python
with open(HG19_VCF_PATH, "r") as f:
    for line in f:
        if line.startswith("#"):
            continue
        row = line.strip().split("\t")
        info_dict = parse_info_field(row[7])
        if set(info_dict.keys()) != set(["OriginalContig", "OriginalStart"]):
            print(row)
```
